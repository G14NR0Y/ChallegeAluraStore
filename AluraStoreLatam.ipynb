### Importación de datos



```python
import pandas as pd

url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)
```

#1. Análisis de facturación



```python
tienda1['Tienda'] = 'Tienda 1'
tienda2['Tienda'] = 'Tienda 2'
tienda3['Tienda'] = 'Tienda 3'
tienda4['Tienda'] = 'Tienda 4'
tiendas = pd.concat([tienda1, tienda2, tienda3, tienda4], ignore_index=True)
facturacion_por_tienda = tiendas.groupby('Tienda')['Precio'].sum().reset_index(name='Facturación Total')
print(facturacion_por_tienda)
```

# 2. Ventas por categoría

```python
ventas_por_categoria_tienda = tiendas.groupby(['Tienda', 'Categoría del Producto']).size().reset_index(name='Cantidad_Vendida')
print(ventas_por_categoria_tienda)
```

```python
ventas_por_categoria_tienda = ventas_por_categoria_tienda.sort_values(['Tienda', 'Cantidad_Vendida'], ascending=[True, False])
print(ventas_por_categoria_tienda)
```

# 3. Calificación promedio de la tienda


```python
promedio_calificacion = tiendas.groupby('Tienda')['Calificación'].mean().reset_index(name='Promedio_Calificación')
print (promedio_calificacion)
```

# 4. Productos más y menos vendidos

```python
producto_mas_menos_vendido_tienda = tiendas.groupby(['Tienda', 'Producto']).size().reset_index(name='Producto_Vendido')
producto_mas_menos_vendido_tienda = producto_mas_menos_vendido_tienda.sort_values(['Tienda', 'Producto_Vendido'], ascending=[True, False])
print(producto_mas_menos_vendido_tienda)
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
print(producto_mas_menos_vendido_tienda)
```

# 5. Envío promedio por tienda

```python
promedio_envio_tienda = tiendas.groupby('Tienda')['Costo de envío'].mean().reset_index(name='Promedio_Envío')
print (promedio_envio_tienda)
```

**graficos**

```python
import matplotlib.pyplot as plt
```

**Facturacion por tienda**

```python
tiendas_nombres = facturacion_por_tienda['Tienda']
facturacion = facturacion_por_tienda['Facturación Total']
plt.figure(figsize=(8, 5))
plt.bar(tiendas_nombres, facturacion, color='skyblue')
plt.title('Facturación Total por Tienda', fontsize=14)
plt.xlabel('Tiendas', fontsize=12)
plt.ylabel('Facturación (S/)', fontsize=12)
plt.show()
```

**producto mas y  menos vendido**

```python
pivot_mas_y_menos_vendido = producto_mas_menos_vendido_tienda.pivot(index='Producto', columns='Tienda', values='Producto_Vendido'

pivot_mas_y_menos_vendido.plot(kind='bar', figsize=(10, 6))
plt.title('Productos mas y menos vendidos por tienda', fontsize=14)
plt.xlabel('Producto', fontsize=12)
plt.ylabel('Cantidad Vendida', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.legend(title='Tiendas')
plt.show()
```

**envio promedio por tienda**

```python
tiendas_nombres = promedio_envio_tienda['Tienda']
Promedio_envio= promedio_envio_tienda['Promedio_Envío']
plt.figure(figsize=(8, 5))
plt.pie(Promedio_envio, labels=tiendas_nombres, autopct='%1.1f%%', startangle=90)
plt.title('Promedio costo de envio', fontsize=14)
plt.xlabel('Tiendas', fontsize=12)
plt.ylabel('Promedio_envio(S/)', fontsize=12)
plt.show()
```

**Informe de Análisis Comparativo de Tiendas**


Este análisis busca comparar el rendimiento de las cuatro tiendas mediante cinco indicadores estratégicos: facturación total, promedio de envío, calificación promedio, volumen de ventas por categoría y número de productos vendidos. Con base en los resultados, se recomendará cuál tienda cerrar y dónde reinvertir para maximizar la rentabilidad.


El análisis inicia con la facturación total, donde la Tienda 1 lidera con más de 1140 millonesde soles, seguida de cerca por las Tiendas 2 y 3, que mantienen cifras competitivas. En contraste, la Tienda 4 registra el menor nivel, con aproximadamente 1038 millones, lo que representa un claro indicador de bajo rendimiento. En cuanto al promedio de envío, la Tienda 4 se posiciona como la más eficiente, presentando los costos más bajos, seguida por las Tiendas 3 y 2, mientras que la Tienda 1 muestra los valores más altos. No obstante, esta ventaja logística no compensa sus limitaciones en ventas. Respecto a la calificación promedio, todas las tiendas mantienen un buen nivel, cercano a cuatro puntos. Sin embargo, la Tienda 3 destaca ligeramente, mientras que la Tienda 4 ocupa la tercera posición, sin mostrar un desempeño sobresaliente. En el análisis de ventas por categoría, las Tiendas 1 y 2 concentran la mayor participación en segmentos estratégicos, como electrónicos y electrodomésticos. La Tienda 3 mantiene un rendimiento intermedio, mientras que la Tienda 4 se rezaga con menores volúmenes en estas categorías clave. Por último, la diversidad de productos vendidos evidencia que la Tienda 1 cuenta con el portafolio más amplio y dinámico, seguida por la Tienda 2. La Tienda 3 mantiene una oferta equilibrada, mientras que la Tienda 4 refleja menor variedad, lo que reduce su atractivo y limita su capacidad para generar mayores ingresos. El análisis integral permite concluir que la Tienda 4, aunque eficiente en costos de envío y con una calificación aceptable, no logra sobresalir en ninguno de los cinco indicadores clave. Presenta la facturación más baja, menor participación en categorías estratégicas, un portafolio limitado y una valoración que no la diferencia frente a la competencia. Por otro lado, la Tienda 1 se mantiene como líder en ingresos y la Tienda 3 destaca por su excelente calificación y un desempeño estable en el resto de indicadores, lo que evidencia un mayor potencial de crecimiento.

Con base en los resultados obtenidos, se recomienda el cierre de la Tienda 4 y la reasignación de recursos hacia la Tienda 3, que combina un alto nivel de satisfacción del cliente con una facturación competitiva. Esta decisión permitirá fortalecer una unidad de negocio con verdadero potencial y optimizar la estrategia comercial para lograr un crecimiento sostenido.
